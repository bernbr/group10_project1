```python
convert_dict = {"Total": float,
                "Retirement Retired workers": float,
               "Retirement Spouses": float,
              "Retirement Children": float,
               "Survivors Widow(er)s and parents": float,
               "Survivors Children": float,
               "Disability Disabled workers": float,
               "Disability Spouses": float,
               "Disability Children": float,
               "Aged 65 or older Men": float,
               "Aged 65 or older Women" : float}

aob_2015 = aob_2015.astype(convert_dict)
aob_2015 = aob_2015[aob_2015['State or area'].str.contains('Outlying areas')==False]
aob_2015 = aob_2015[aob_2015['State or area'].str.contains('Unknown')==False]
```

```python
#Calculate sums and totals for each column, grouped by state
state_totals = aob_2015.groupby('State or area').sum()['Total']
retirement_workers = aob_2015.groupby('State or area').sum()['Retirement Retired workers']
retirement_spouses = aob_2015.groupby('State or area').sum()['Retirement Spouses']
retirement_children = aob_2015.groupby('State or area').sum()['Retirement Children']
survivors_widows = aob_2015.groupby('State or area').sum()['Survivors Widow(er)s and parents']
survivors_children = aob_2015.groupby('State or area').sum()['Survivors Children']
disability_workers = aob_2015.groupby('State or area').sum()['Disability Disabled workers']
disability_spouses = aob_2015.groupby('State or area').sum()['Disability Spouses']
disability_children = aob_2015.groupby('State or area').sum()['Disability Children']
sixty_plus_men = aob_2015.groupby('State or area').sum()['Aged 65 or older Men']
sixty_plus_women = aob_2015.groupby('State or area').sum()['Aged 65 or older Women']
```

```python
#Create new dataframe
summary_data = {'Total Amount': state_totals,
                'Retired Workers': retirement_workers,
               'Retirement Spouses': retirement_spouses,
               'Retirement Children': retirement_children,
               'Survivors Widow(er)s and Parents': survivors_widows,
               'Survivors Children':survivors_children,
               'Disability Workers':disability_workers,
               'Disability Spouses': disability_spouses,
               'Disability Children': disability_children,
               'Aged 65 or older Men':sixty_plus_men,
               'Aged 65 or older Women': sixty_plus_women}

summary_df = pd.DataFrame(summary_data)
```

```python
#Iterate through indices to get totals into empty lists

retirement_totals = []
survivors_totals = []
disability_totals = []
sixty_five_plus_totals = []

for i, row in summary_df.iterrows():
    retirement_total=summary_df['Retired Workers'] + summary_df['Retirement Spouses'] + summary_df['Retirement Children']
    retirement_totals.append(retirement_total)
    
    survivor_total = summary_df['Survivors Widow(er)s and Parents'] + summary_df['Survivors Children']
    survivors_totals.append(survivor_total)
    
    disability_total = summary_df['Disability Workers'] + summary_df['Disability Children'] + summary_df['Disability Spouses']
    disability_totals.append(disability_total)
    
    sixty_five_plus_total = summary_df['Aged 65 or older Men'] + summary_df['Aged 65 or older Women']
    sixty_five_plus_totals.append(sixty_five_plus_total)
```

```python
#Add columns to dataframe
summary_df['Retirement Totals'] = retirement_totals[1]
summary_df['Survivor Totals'] = survivors_totals[1]
summary_df['Disability Totals'] = disability_totals[1]
summary_df['Aged 65 and older Totals'] = sixty_five_plus_totals[1]
summary_df.head()
```

```python
summary_df.index
```

```python
#Percentage of retirement, survivor, disability, 65 plus out of total
percent_retirement = summary_df['Retirement Totals']/summary_df['Total Amount'] * 100
percent_survivor = summary_df['Survivor Totals']/summary_df['Total Amount'] * 100
percent_disability = summary_df['Disability Totals']/summary_df['Total Amount'] * 100
percent_age = summary_df['Aged 65 and older Totals']/summary_df['Total Amount'] * 100
```

```python
#Create new dataframe for percentages
percent_data = {'Retirement Percentage': percent_retirement,
               'Survivor Percentage':percent_survivor,
               'Disability Percentage': percent_disability,
               '65+ Percentage': percent_age}

percentage_table_2015 = pd.DataFrame(percent_data)
percentage_table_2015.head()
```
